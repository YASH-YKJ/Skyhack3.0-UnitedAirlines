Task 3: Aggregate (Summarize) Your Data


Step 3.1: Aggregate the Passenger Data (pnr_df)


```python
# Define the columns that uniquely identify a flight
flight_identifier = ['company_id', 'flight_number', 'scheduled_departure_date_local']

# Define the columns we want to count up for each flight
passenger_counts = {
    'total_pax': 'sum',
    'lap_child_count': 'sum',
    'is_child': 'sum',
    'is_stroller_user': 'sum'
}

# Group by the flight identifier and perform the aggregation
pnr_summary_df = pnr_df.groupby(flight_identifier).agg(passenger_counts).reset_index()
```

Task 4: Aggregate (Summarize) Your Baggage Data


Step 4.1: Create a Helper Column


```python
bag_df['is_transfer_bag'] = (bag_df['bag_type'] == 'Transfer').astype(int)
```

Step 4.2: Aggregate the Baggage Data


```python
# The flight identifier is the same as before
flight_identifier = ['company_id', 'flight_number', 'scheduled_departure_date_local']

# Define the columns we want to summarize
bag_counts = {
    'bag_tag_unique_number': 'count', # This will count total bags
    'is_transfer_bag': 'sum'       # This will sum up our helper column to count transfer bags
}

# Group by the flight identifier and perform the aggregation
bag_summary_df = bag_df.groupby(flight_identifier).agg(bag_counts).reset_index()

# Let's rename the columns to be more descriptive
bag_summary_df.rename(columns={'bag_tag_unique_number': 'total_bags', 
                               'is_transfer_bag': 'transfer_bags'}, inplace=True)
```

Task 5: Merge All Data into a Master Table


Step 5.1: Merge Flights with Passenger Summary


```python
# The flight identifier is the same as before
flight_identifier = ['company_id', 'flight_number', 'scheduled_departure_date_local']

# Merge flight_df with the passenger summary
master_df = pd.merge(flight_df, pnr_summary_df, on=flight_identifier, how='left')
```

Step 5.2: Merge the Result with Baggage Summary


```python
# Merge the master_df with the baggage summary
master_df = pd.merge(master_df, bag_summary_df, on=flight_identifier, how='left')
```

Step 5.3: Clean Up Missing Values


```python
# List of new columns we added
new_count_columns = ['total_pax', 'lap_child_count', 'is_child', 'is_stroller_user', 'total_bags', 'transfer_bags']

# Fill any blanks in these columns with 0
master_df[new_count_columns] = master_df[new_count_columns].fillna(0)
```

Task 6: Feature Engineering (Creating Smarter Clues)


Step 6.1: Calculate Seat Fullness (load_factor)


```python
master_df['load_factor'] = master_df['total_pax'] / master_df['total_seats']
```

Step 6.2: Calculate Time Pressure (turn_time_pressure)


```python
master_df['turn_time_pressure'] = master_df['scheduled_ground_time_minutes'] - master_df['minimum_turn_minutes']
```

Step 6.3: Calculate Bag Complexity (transfer_bag_ratio)


```python
# --- Fully Corrected Code for transfer_bag_ratio ---

# Calculate the ratio. This might create NaN and 'inf' values.
master_df['transfer_bag_ratio'] = master_df['transfer_bags'] / master_df['total_bags']

# The safe way to clean up the column:
# 1. Replace all infinity values with NaN (blanks).
# 2. Fill all NaN values (blanks) with 0.
# 3. Assign the cleaned result back to the column.
master_df['transfer_bag_ratio'] = master_df['transfer_bag_ratio'].replace([np.inf, -np.inf], np.nan).fillna(0)
```

Task 9: Building the Flight Difficulty Score (Deliverable 2)

Step 9.1: Select and Normalize Features


```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# --- Step 1 & 2: Define features and the custom scaling function (unchanged) ---
master_df['turn_time_pressure_inv'] = master_df['turn_time_pressure'] * -1
features_to_scale = [
    'load_factor',
    'turn_time_pressure_inv',
    'transfer_bag_ratio',
    'special_requests_count'
]

def scale_daily_features(group):
    scaler = MinMaxScaler()
    if len(group) < 2:
        scaled_data = np.zeros((len(group), len(features_to_scale)))
    else:
        scaled_data = scaler.fit_transform(group[features_to_scale])
    return pd.DataFrame(
        scaled_data,
        index=group.index,
        columns=['load_factor_scaled', 'ttp_inv_scaled', 'tbr_scaled', 'src_scaled']
    )

# --- Step 3: Apply our function (this creates the complex index) ---
scaled_features_df = master_df.groupby('scheduled_departure_date_local').apply(scale_daily_features)

# --- THE FIX: Reset the index to remove the extra level created by groupby ---
scaled_features_df = scaled_features_df.reset_index(level=0, drop=True)

# --- Step 4: Join the new scaled columns back to the master_df (will now work) ---
master_df = master_df.join(scaled_features_df)

print("Features have been scaled correctly with the definitive .apply() method.")
```

```python
# --- Step 3: Define weights (they must sum to 1.0) ---
weights = {
    'ttp_inv_scaled': 0.40,      # 40% importance
    'load_factor_scaled': 0.25,  # 25% importance
    'src_scaled': 0.20,          # 20% importance
    'tbr_scaled': 0.15           # 15% importance
}

# --- Step 4: Calculate the final score ---
master_df['difficulty_score'] = (
    master_df['ttp_inv_scaled'] * weights['ttp_inv_scaled'] +
    master_df['load_factor_scaled'] * weights['load_factor_scaled'] +
    master_df['src_scaled'] * weights['src_scaled'] +
    master_df['tbr_scaled'] * weights['tbr_scaled']
)

print("Difficulty Score calculated.")
```

```python
# --- Step 5: Define the quantile boundaries ---
quantiles = master_df['difficulty_score'].quantile([0, 0.3, 0.8, 1.0])

# --- Step 6: Create the classification labels ---
labels = ['Easy', 'Medium', 'Difficult']
master_df['difficulty_class'] = pd.cut(
    master_df['difficulty_score'],
    bins=quantiles.values,
    labels=labels,
    include_lowest=True # Include the lowest value in the first bin
)

print("Flights have been classified.")
```
